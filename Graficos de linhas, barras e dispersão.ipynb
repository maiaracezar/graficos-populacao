```
import matplotlib.pyplot as plt

# Dados de população por região em 2020 (em milhões)
regioes = ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
populacao_2020 = [17.9, 55.2, 84.1, 29.0, 16.9]

# Criando o gráfico de barras
plt.figure(figsize=(10,6))
plt.bar(regioes, populacao_2020, color='skyblue')
plt.title('População por Região em 2020')
plt.xlabel('Regiões')
plt.ylabel('População (milhões)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd

# Dados de população por região ao longo dos anos (em milhões)
dados = {
    'Ano': [2000, 2005, 2010, 2015, 2020],
    'Norte': [12.5, 13.8, 15.2, 16.5, 17.9],
    'Nordeste': [47.1, 49.3, 51.5, 53.7, 55.2],
    'Sudeste': [72.3, 75.1, 78.0, 81.4, 84.1],
    'Sul': [24.2, 25.4, 26.6, 27.9, 29.0],
    'Centro-Oeste': [11.8, 13.0, 14.4, 15.7, 16.9]
}

# Criando DataFrame
df = pd.DataFrame(dados)

# Plotando o gráfico de linhas
plt.figure(figsize=(10,6))
for regiao in ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']:
    plt.plot(df['Ano'], df[regiao], marker='o', label=regiao)

plt.title('Evolução da População por Região (2000–2020)')
plt.xlabel('Ano')
plt.ylabel('População (milhões)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt

# Dados fictícios de população e densidade por estado
estados = ['SP', 'RJ', 'MG', 'BA', 'RS']
populacao = [44.0, 17.0, 21.0, 15.0, 11.0]  # em milhões
densidade = [166, 365, 36, 25, 39]  # hab/km²

# Scatterplot simples
plt.figure(figsize=(10,6))
plt.scatter(densidade, populacao, color='green')
for i, estado in enumerate(estados):
    plt.text(densidade[i]+2, populacao[i], estado)
plt.title('População vs. Densidade Demográfica por Estado')
plt.xlabel('Densidade (hab/km²)')
plt.ylabel('População (milhões)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Bubbleplot com tamanho proporcional à população
plt.figure(figsize=(10,6))
tamanhos = [p*10 for p in populacao]  # Ajustando o tamanho dos pontos
plt.scatter(densidade, populacao, s=tamanhos, alpha=0.5, color='green')
for i, estado in enumerate(estados):
    plt.text(densidade[i]+2, populacao[i], estado)
plt.title('População vs. Densidade Demográfica por Estado (Bubbleplot)')
plt.xlabel('Densidade (hab/km²)')
plt.ylabel('População (milhões)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
